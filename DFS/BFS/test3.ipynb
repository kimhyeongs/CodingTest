스택(stack) -> 선입후출(박스 쌓기) </br>
큐(que) -> 선입선출(deque 자료 구조 사용) </br>
재귀 함수 -> 자기 자신을 호출(종료 조건 필수) </br>
완전 탐색 -> BFS </br>
부분 탐색 -> DFS </br>
Best first search (우선순위 큐 이용) > Depth first search (스택 이용) > Breath first search (큐 이용) </br>

### DFS에 관한 얘기 </br>
DFS는 스택 자료구조를 이용하며, 재귀를 이용한다. </br>


### BFS에 관한 얘기 </br>
BFS는 큐 자료구조를 이용하며, 반복문을 이용한다. </br>
특정 조건에서의 최단 경로를 구할 때 많이 사용 -> ex) weight가 모두 동일할 떄 


주로 문제에서 최단거리를 구하라고하면 DFS보단 BFS로 구해야 한다. </br>
파이썬을 이용할 때에는 deque를 사용하여 구하는 것이 좋다. </br>
리스트랑 차이가 없어 보이지만, 파이썬에서 deque가 리스트보다 시간 면에서 빠르기 때문이다. </br>

> ### DFS </br>
>> -> 깊이 우선 탐색(노드와 노드의 관계 표현), 시간복잡도 O(N)


>표현 방식:
  >>1. 인접 행렬 -> 행렬을 통해 표현 
      >>>ex) [[0,7,5],[7,0,INF],[5,INF,0]]
  >>2. 인접 리스트 -> 리스트를 통해 표현
      >>>ex) [0] [(1,7),(2,5)]
             [1] [(0,7)]
             [2] [(0,5)] 
      

>탐색 방식:
>>1. 탐색 시작 노드를 스택에 삽입 후 방문 처리
>>2. 방문하지 않은 노드를 스택에 삽입, 방문하지 않은 노드가 주변에 없으면 스택 최상단 노드를 꺼냄
>>3. 반복 

```python
# DFS 기본 구현
def dfs(graph, v, visited):
    # 현재 노드를 방문 처리
    visited[v] = True
    print(v,end=' ')
    # 현재 노드와 연결된 다른 노드를 재귀적으로 방문
    for i in graph[v]:
        if not visited[i]:
            dfs(graph, i, visited)

graph = [
    [], # 0인 노드를 편하게 표현
    [2, 3, 8],
    [1, 7],
    [1, 4, 5],
    [3, 5],
    [3, 4],
    [7],
    [2, 6, 8],
    [1, 7]
]

# visited 초기화
visited = [False] * 9

dfs(graph, 1, visited)
```

> ### 음료수 얼려 먹기 ( P.149 ) - 복습 필요 
>>- 상하좌우 연결된 모든 부분을 탐색하여 0 -> 1 로 변경하여 반복적으로 세는 경우가 없도록 만듬
>>- 부분 탐색이므로 DFS 사용
>>- 범위를 넘어가는 경우에는 False 처리

```python
n,m = map(int, input().split())

graph = []
for i in range(n):
    graph.append(list(map(int,input().split())))

def dfs(x,y):
    if x <= -1 and x >= n and y <= -1 and y >= m:
        return False
    
    if graph[x][y] == 0:
        # 시작 노드 방문 처리
        graph[x][y] = 1
        # 상하좌우 모두 연결된 곳 방문 처리
        dfs(x-1,y)
        dfs(x,y-1)
        dfs(x+1,y)
        dfs(x,y+1)
        return True
    else:
        return False
    
result = 0
for i in range(n):
    for j in range(m):
        if dfs(i,j) == True:
            result+=1

print(result)
    
```

> ### BFS 
>> -> 너비 우선 탐색

>탐색 방식:
>>1. 탐색 시작 노드를 큐에 삽입하고 방문 처리
>>2. 큐에서 노드를 꺼내 해당 노드의 인접 노드 중에서 방문하지 않은 노드를 모두 큐에 쌓음
>>3. 반복

- popleft() : 큐에서 가장 처음에 입력된 노드를 꺼내고 삭제 </br>
- 지나온 위치 값을 변환하여 다시 안가게 하는 방식을 많이 사용 </br>
- popleft()를 통해 완전 탐색 진행 </br> 

```python
from collections import deque

def bfs(graph , start , visited):
    #큐 구현
    queue = deque([start])
    
    #현 노드 방문처리
    visited[start] = True
    
    #큐 빌때까지 반복함
    while queue:
        v = queue.popleft()
        print(v,end=" ")

        #아직 방문하지 않은 원소들 큐에 삽입함
        for i in graph[v]:
            if not visited[i]:
                queue.append(i)
                visited[i] = True

graph = [
    #보통 첫 번째를 1이라고 하기 떄문에, 0번쨰 노드 아예 비워둠둠
    [],
    [2,3,8],
    [1,7],#2번 노드와 인접
    [1,4,5], #3번 노드와 인접...
    [3,5],
    [3,4],
    [7],
    [2,6,8],
    [1,7]
]
#각 노드의 방문 정보를 표현하기 위한 리스트
visited = [False]*9
#정의된 bps 호출
```

> ### 미로 탈출 ( P.152 )

>>- 완전 탐색 문제이므로 bfs 사용
>>- 큐를 이용하여 완전 탐색 진행
>>- 지나간 곳의 graph 값을 지나온 횟수로 바꿈

```python
from collections import deque

n,m = map(int, input().split())

graph = []
for i in range(n):
    graph.append(list(map(int, input())))

dx = [-1, 1, 0, 0]
dy = [0, 0, -1, 1]

def bfs(x,y):
    # 큐 구현
    queue = deque()
    queue.append((x,y))

    while queue:
        # 큐에서 위치 꺼내기
        x,y = queue.popleft()

        # 상하좌우 모두 확인
        for i in range(4):
            nx = x + dx[i]
            ny = y + dy[i]

            if nx < 0 or nx >= n or ny < 0 or ny >= m:
                continue
            
            if graph[nx][ny] == 0:
                continue

            if graph[nx][ny] == 1:
                graph[nx][ny] == graph[x][y]+1
                queue.append((nx,ny))
    
    return graph[n-1][m-1]
```
